```python
import pandas as pd
import numpy as np

import os
import json
import requests
from tqdm import tqdm
```

```python
META_FILENAME = '/Volumes/Education/TAMU/Spring_23/CS670_Info_Storage_Retrieval/Project/Data/meta_AMAZON_FASHION.json'
IMAGE_FOLDER_LR = './Data/Image_Data/Low_Res'
IMAGE_FOLDER_HR = './Data/Image_Data/High_Res'
```

```python

with open(META_FILENAME,'r') as file:
    data = file.readlines()
    

# Create a dataframe from the data
    

data = [json.loads(row) for row in data] # Convert each element from string to dictionary
data_df = pd.DataFrame(data, columns = ['title', 'feature', 'brand', 'asin', 'imageURLHighRes', 'imageURL', 'rank', 'date'])
data_df.sample(5)
```

```python
if not os.path.exists(IMAGE_FOLDER_LR):
    os.makedirs(IMAGE_FOLDER_LR)
    
if not os.path.exists(IMAGE_FOLDER_HR):
    os.makedirs(IMAGE_FOLDER_HR)
```

```python
image_dictionary = {}
success_count = 0
failed_count = 0
data_na = 0

# starting_index = 42780
starting_index = 74320
ending_index = 80000

starting_index = 170000
# ending_index = 80000

for index, product in tqdm(data_df.iloc[starting_index:, :].iterrows(), total = data_df.shape[0], initial = starting_index):
        asin = product['asin']
        hr_urls = product['imageURLHighRes']
        sucessful_extractions = []

        if hr_urls is np.nan:
            data_na +=1
            continue

        for ind, url in enumerate(hr_urls):
            try:
                extension = url.split('.')[-1]
                r = requests.get(url, allow_redirects=True)
                image_save_fname = asin + '_' + str(ind) +'.' + extension
                open(os.path.join(IMAGE_FOLDER_HR,image_save_fname), 'wb').write(r.content)
                sucessful_extractions.append(image_save_fname)
                success_count += 1
            except:
                failed_count += 1

            image_dictionary[asin] = sucessful_extractions
```
